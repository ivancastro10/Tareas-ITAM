```python
metadata_file = "../files/lec02/world-bank/metadata.csv"
```

```python
!head -n 5 ../files/lec02/world-bank/metadata.csv
```

```python
metadata = pd.read_csv(metadata_file, index_col=0).iloc[:, :-1]
metadata.head(3)
```

```python
gdp.head(3)
```

```python
gdp_merged = metadata.join(gdp, how="inner")
```

```python
gdp_merged.head()
```

```python
gdp_merged.groupby("IncomeGroup").sum().T.plot()
```

```python
(gdp_merged.groupby("IncomeGroup").sum()
           .drop("High income").T.plot())
```

```python
res_income = gdp_merged.groupby("IncomeGroup").sum().T
```

```python
l1 = [1, 2, 3]
l2 = ["a", "b", "c"]
```

```python
fig, ax = plt.subplots(2, 2)
ax = ax.ravel()
for axi, c in zip(ax, res_income):
    res_income[c].plot(ax=axi)
    axi.set_title(c)
plt.tight_layout()
```
