```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn, seed
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import f1_score
```

```python
noise = 0.2
X, y = make_moons(200, noise=noise, random_state=2718)
X_test, y_test = make_moons(40, noise=noise, random_state=31415)
plt.scatter(*X.T, c=y, cmap="Set2")
```

```python
## 1) Crea un pipeline llamado model que primero transforme el espacio con un polinomio de grado 30 (sin incluir p0) y 
## posteriormente considere la transformación de los datos con una regresión logística sin regularización 
## (LogisticRegression(penalty="none"))

model = Pipeline([
    ("poli", PolynomialFeatures(degree=30, include_bias = False)),
    ("log_reg", LogisticRegression(penalty="none"))
])
```

```python
## 2)Realiza un Bagging Classifer de model llamado bagging_logreg con n_estimators=100 y random_state=314

bagging_logreg = BaggingClassifier(model, n_estimators=100, random_state=314)
```

```python
## 3) Entrena model y bagging_logreg considerando X y y definido arriba

bagging_logreg.fit(X,y)
```

```python
from matplotlib.colors import ListedColormap

def plot_classf_model(X, y, model, ax, title=None, step=0.01, epsilon=0.1, contour=False):
    """
    Contourplot a two-class underlying model, and test datapoints.
    
    Paramters
    ---------
    X: Matrix(mxn, R)
        Matrix of feature vectors with examples
    y: n-dimensional vector 
        Actual classes for each of the 'm' provided examples.
    """
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    points_cmap = ListedColormap(["Yellow", "Blue"])
    x0, x1 = X[:,0].min() - epsilon, X[:,0].max() + epsilon
    y0, y1 = X[:,1].min() - epsilon, X[:,1].max() + epsilon
    
    xx, yy = np.mgrid[x0:x1:step, y0:y1:step]
    
    preds = model.predict(np.c_[xx.ravel(), yy.ravel()])
    preds = preds.reshape(xx.shape)
    
    if contour:
        ax.contour(xx, yy, preds)
    ax.pcolormesh(xx, yy, preds, cmap=custom_cmap)
    ax.scatter(*X.T, c=y, s=50, edgecolor="k", cmap=points_cmap)
    if title is not None:
        ax.set_title(title)
```

```python
## 4) Grafica la función de decisión dada por model y por bagging_logreg ¿Qué observas?
## Grafica usando bagging_logreg
fig, ax = plt.subplots()
plot_classf_model(X, y, bagging_logreg, ax)
```

```python
model.fit(X,y)
```

```python
## Grafica usando unicamente model
fig, ax = plt.subplots()
plot_classf_model(X, y, model, ax)
```

```python
## 5) Considerando un f1_score (dentro de sklearn.metrics), calcula el f1-score sobre el test set (X_test, y_test) 
##  considerando model y bagging_logreg. ¿Qué resultados arroja?

ytest_score_model = f1_score(y_test, model.predict(X_test))
ytest_score_bagg = f1_score(y_test, bagging_logreg.predict(X_test))

print(f"El score usando model:{ytest_score_model:0.2f}")
print(f"El score usando bagging_logreg:{ytest_score_bagg:0.2f}")
```

Al final sacando el score sobre los datos de prueba, vemos como hay una notable diferencia entre usar el bagging y utilizar unicamente model (regresión logistica) una vez. 
Para bagging, la prueba f1 score, que es un promedio ponderado entre la precisión y el recall, queda en 0.95 muy cercano a su mejor puntaje que sería de 1. En cambio para model, con una sola regresión logistica, el score es de 0.76. 
También se puede observar desde las gráficas como para el bagging, que a pesar de siempre usar el mismo modelo, va tomando diferentes muestras y llega a separar de una mejor manera los datos.
